```python
import nltk
import random
from nltk.corpus import movie_reviews
```

```python
nltk.download('movie_reviews')
```

```python
from nltk.corpus import movie_reviews
```

```python
documents = [(list(movie_reviews.words(fileid)), category)
             for category in movie_reviews.categories()
             for fileid in movie_reviews.fileids(category)]
```

```python
random.shuffle(documents)
```

```python
print(documents[1])
```

```python
all_words=[]
for w in movie_reviews.words():
    all_words.append(w.lower())
all_words=nltk.FreqDist(all_words)
```

```python
print(all_words.most_common(15))
```

```python
print(all_words["stupid"])
```

```python
#We Want words as features
#These words can help us understand whether the review is positive or negative
#For Example:stupid-Negative,Good-Positive
```

```python
print(all_words["good"])
```

```python
word_features=list(all_words.keys())[:3000]
```

```python
def find_features(document):
    words=set(document)
    features={}
    for w in word_features:
        features[w]=(w in words)
    return features
```

```python
print((find_features(movie_reviews.words('neg/cv000_29416.txt'))))
```

```python
featuresets = [(find_features(rev), category) for (rev, category) in documents]
```

```python
#So,we set the most common words as the features and then we will use these features to classify using the naive bayes classifier
```

```python
featuresets[100]
```

```python
train_set=featuresets[:1900]
test_set=featuresets[1900:]
```

```python
classifier=nltk.NaiveBayesClassifier.train(train_set)
```

```python
print("Classifier accuracy percent:",(nltk.classify.accuracy(classifier, test_set))*100)
```

```python
classifier.show_most_informative_features(15)
```
